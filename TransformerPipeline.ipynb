```python
#To run transformers we would need GPU. Hence, we can you `https://colab.research.google.com/` with free T4 Runtime and execute below code.

#Log-in to HuggingFace.co and generate write TOKEN and put it in colab secret


!pip install -q --upgrade torch==2.5.1+cu124 torchvision==0.20.1+cu124 torchaudio==2.5.1+cu124 --index-url https://download.pytorch.org/whl/cu124

!pip install -q diffusers transformers==4.48.3 

from huggingface_hub import login
from transformers import pipeline
import torch
from diffusers import DiffusionPipeline
from google.colab import userdata


hf_token=userdata.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)

print("\n\n *****Sentiment Analysis***** \n\n")
sentiment = pipeline("sentiment-analysis", device="cuda")
result = sentiment("I am excited to learn LLM and AI")
print(result)

print("\n\n *****Question Answer***** \n\n")
question_answer = pipeline("question-answering", device="cuda")
result = question_answer(question="What is John's age?", context="Hi, My name is John and I am 18 years old.")
print(result)

print("\n\n *****Translation_en_to_fr***** \n\n")
translator = pipeline("translation_en_to_fr", device="cuda")
result = translator("I am excited to learn LLM and AI")
print(result[0]['translation_text'])

print("\n\n *****Text Generation***** \n\n")
generator = pipeline("text-generation", device="cuda")
result = generator("Complete this poem. Johnny Johnny Yes Papa Eating")
print(result)

print("\n\n *****Image Generation***** \n\n")
image_gen = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
    ).to("cuda")

text_prompt = "Waterfall at mountains and sun is rising."     
image = image_gen(prompt=text_prompt).images[0]
image
```
